```python
# Transformaciones
# 1. Unir invoice_header con invoice_products
invoice_details = pd.merge(invoice_header, invoice_products, on='invoice_id')

# 2. Calcular el importe total en la moneda original
invoice_details['total_amount'] = invoice_details['quantity'] * invoice_details['price']

# 3. Convertir a euros usando el tipo de cambio
invoice_details = pd.merge(invoice_details, daily_currencies, left_on=['currency', 'invoice_date'], right_on=['currency', 'date'])
invoice_details['total_amount_eur'] = invoice_details['total_amount'] * invoice_details['exchange_rate']

# 4. Calcular lead time real
invoice_details['lead_time_real'] = (pd.to_datetime(invoice_details['reception_date']) - pd.to_datetime(invoice_details['order_date'])).dt.days

# 5. Asignar lead time teórico
def get_theoretical_lead_time(country):
    if country == 'Spain':
        return 10
    elif country in ['France', 'Germany', 'Italy']:  # Ejemplo de países europeos
        return 20
    else:
        return 45

invoice_details['lead_time_theoretical'] = invoice_details['supplier_country'].apply(get_theoretical_lead_time)

# 6. Agregaciones y creación del datamart
# Ejemplo: Compras diarias por sección
daily_purchases_by_section = invoice_details.groupby(['invoice_date', 'section'])['total_amount_eur'].sum().reset_index()

# Guardar el datamart
daily_purchases_by_section.to_csv('daily_purchases_by_section.csv', index=False)

# Continuar con el resto de transformaciones y guardar las tablas del datamart
```
